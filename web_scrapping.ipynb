Detik.com

```python
hal = 1
teks =[]
tag =['politik', 'partai', 'pemerintah']

urllib3.disable_warnings()

while hal <=20:
  for i in tag:
    link = f"https://www.detik.com/tag/{i}/?sortby=time&page={hal}"
    r = url.request("GET", link)
    s = BeautifulSoup(r.data, 'html.parser')
    for i in s.select('article a span p'):
            teks.append(i.get_text(strip=True))
  hal +=1
```

```python
labelberita = []
for i in range(len(teks)):
    labelberita.append('valid')

detik = pd.DataFrame(teks,columns=['judul'])
detik['label_berita'] = labelberita
```

```python
import requests
from bs4 import BeautifulSoup

# tags yang ingin di-scrape
tags = ['politik', 'partai', 'pemerintah','pemilu']

# list untuk menyimpan judul berita
teks = []

# loop untuk setiap tag
for tag in tags:
    # loop untuk halaman 1-10
    for page in range(1, 2):
        # URL untuk halaman dengan tag dan nomor halaman
        url = f'https://www.cnnindonesia.com/tag/{tag}/{page}'

        # lakukan request halaman menggunakan requests library
        r = requests.get(url)

        # parsing HTML menggunakan BeautifulSoup
        soup = BeautifulSoup(r.content, 'html.parser')

        # temukan semua judul berita pada halaman tersebut
        judul_berita = soup.find_all('h2', {'class': 'title'})

        # tambahkan judul berita ke dalam list teks
        for judul in judul_berita:
            teks.append(judul.text.strip())

# tampilkan judul berita
# for judul in teks:
#     print(judul)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

tags = ['politik','kpu','dpr','jokowi','presiden', 'partai', 'pemerintah','pemilu']
pages = 20
data = []

for tag in tags:
    for page in range(1, pages+1):
        url = f'https://www.detik.com/tag/{tag}/?sortby=time&page={page}'
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        articles = soup.find_all('article')

        for article in articles:
            link = article.find('a')['href']
            if 'https://www.detik.com/' in link:
                response = requests.get(link)
                soup = BeautifulSoup(response.content, 'html.parser')
                title = soup.find('h1', class_='detail__title').text.strip()
                content = soup.find('div', class_='detail__body-text').text.strip().split('\n')[0]
                data.append({'Judul': title, 'Isi Berita': content})

df = pd.DataFrame(data)
df.to_excel('hasil_scraping_detik.xlsx', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

tags = ['politik','kpu','dpr','jokowi','presiden', 'partai', 'pemerintah','pemilu']
pages = 5
data = []
urls = []

for tag in tags:
    for page in range(1, pages+1):
        url = f'https://www.kompas.com/tag/{tag}?page={page}'
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")
        articles = soup.find_all("article")
        for article in articles:
            link = article.find("a")["href"]
            if "https://www.kompas.com/" in link:
                urls.append(link)

for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    title = soup.find("h1", class_="read__title")
    if title is not None:
        title = title.text.strip()
    paragraphs = soup.find_all("div", class_="detail__body-text itp_bodycontent")
    content = ''
    for paragraph in paragraphs:
        content += paragraph.text.strip() + '\n'
    data.append({"Judul": title, "Isi Berita": content})

df = pd.DataFrame(data)
df.to_excel("hasil_scraping_kompas.xlsx", index=False)
```
